```python
def clss(x):
  if x == 'Fog':
    return 1
  if x == 'Clear':
    return 1
  if x == 'Rain':
    return 1
  if x == 'Snow':
    return 1
  if x == 'Fog, Snow':
    return 2
  if x == 'Rain, Snow':
    return 2
  if x == 'Clear, Rain':
    return 2
  if x == 'Clear, Fog':
    return 2
  if x == 'Fog, Rain':
    return 2
  if x == 'Clear, Snow':
    return 2
  if x == 'Clear, Fog, Rain, Snow':
    return 4
  if x == 'Clear, Fog, Snow':
    return 3
  if x == 'Clear, Fog, Rain':
    return 3
  if x == 'Fog, Rain, Snow':
    return 3
  if x == 'Clear, Rain, Snow':
    return 3
  else:
    return 0
```

```python
df_3c['coded'] = df_3c['weather_conditions'].apply(clss)
print(df_3c['coded'].sum())
df_3c[['coded','weather_conditions']]
```

```python
import statsmodels.api as sm
Y = df_3c['pedestrian_incidents']
X = sm.add_constant(df_3c['temperature'])
model = sm.OLS(Y, X).fit()
print(round(model.params, 2))
#A=model.params[1]-model.params[0]
#round(A,2)
```

Вопрос 24

```python
Y_pr = model.predict(X)
MAE = sum(abs(Y-Y_pr))/len(Y)
round(MAE,2)
```
