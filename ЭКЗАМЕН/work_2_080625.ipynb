Вопрос 23

```python
import statsmodels.api as sm
Y = df_2c['price']
X = sm.add_constant(df_2c['sqft_living'])
model = sm.OLS(Y, X).fit()
print(round(model.params, 2))
#A=model.params[1]-model.params[0]
#round(A,2)
```

Вопрос 24

```python
Y_pr = model.predict(X)
MSE = sum((Y-Y_pr)**2)/len(Y)
round(MSE,2)
```
