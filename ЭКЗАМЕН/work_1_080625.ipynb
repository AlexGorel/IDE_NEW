Вопрос 23

```python
import statsmodels.api as sm
Y = df_1c['Price']
X = sm.add_constant(df_1c['Mileage'])
model = sm.OLS(Y, X).fit()
res=model.params[0]*model.params[1]
round(res,2)
print(model.params) 
print(round(res,2))
```

Вопрос 24

```python
Y_pr = model.predict(X)
MSE = sum((Y-Y_pr)**2)/len(Y)
round(MSE, 2)
```
