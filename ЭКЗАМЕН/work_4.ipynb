вопрос 20

```python
def findLen(str):
    counter = 0
    while str[counter:]:
        counter += 1
    return counter
```

```python
df_4c['coded'] = df_4c['song_name'].apply(findLen)
round(df_4c['coded'].mean(),2)
```

вопрос 21

```python
avg = df_4c['tempo'].mean()
std = df_4c['tempo'].std()
low2 = avg-2.5*std #можно посчитать отклонение и на одно, и на два, и на три
up2 = avg+2.5*std
low2, up2
```

```python
df_4c[(df_4c['tempo']<low2) | (df_4c['tempo']>up2)].shape[0]
```

Вопрос 23

```python
import statsmodels.api as sm
Y = df_4c['popularity']
X = sm.add_constant(df_4c['danceability'])
model = sm.OLS(Y, X).fit()
print(round(model.params, 2))
A=model.params[1]-model.params[0]
round(A,2)
```

Вопрос 24

```python
round(model.rsquared,2)
```
