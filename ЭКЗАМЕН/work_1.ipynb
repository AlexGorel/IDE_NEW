Вопрос 23

```python
import statsmodels.api as sm
Y = df_5c['DayPassPriceAdult']
X = sm.add_constant(df_5c['TotalSlope'])
model = sm.OLS(Y, X).fit()
round(model.params, 2)[1]
```

Вопрос 24

```python
Y_pr = model.predict(X)
MAE = sum(abs(Y-Y_pr))/len(Y)
round(MAE,2)
```
