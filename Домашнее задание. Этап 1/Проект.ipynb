Для 'Victim Age' найдем выбросы методом N*std

ЛИНЕЙНАЯ РЕГРЕСИЯ

Мы хотим построить модель, которая показывает как ВОЗРАСТ ЖЕРТВЫ (**Victim Age**) оказывает влияение на КОЛИЧЕСТВО ПРЕСТУПЛЕНИЙ (**Incident**)

```python
fig, ax = plt.subplots(figsize=[10,8])
ax.scatter(x = data_clean['Incident'], y = data_clean['Victim Age'], alpha=0.7, c = 'purple')
ax.set_title('Зависимость количества преступлений от возраста жертвы', fontsize=15)
ax.set_xlabel('Количество преступлений', fontsize=12)
ax.set_ylabel('Возраст жертвы', fontsize=12)
plt.show()
```

```python
import statsmodels.api as sm

X = sm.add_constant(data_clean[['Incident']])  # предиктор/независимая переменная
Y = data_clean['Victim Age']                                       # целевая переменная/зависимая переменная
model = sm.OLS(Y, X).fit()  
```

```python
model.params
```

w0=33.467622. Если нет совсем проишествий, то возраст жерт будет 33,46 лет

w1=-0.001030 при увеличении количества проишествий на единицу возраст жертвы будет уменьшаться на 0,001030 лет

#### Коэфициент детерминации ($R^2$)
Показывает долю данных, которые объясняет модель:

- Меняется от 0 до 1
- Чем ближе к 1, тем лучше


```python
model.rsquared
```

плохое качество модели. Необходимо иметь больше признаков

**Среднеквадратичная ошибка**:

$$ MSE = \frac{\Sigma(Y-\widehat{Y})^2}{n} $$

$\Sigma$ - знак суммы, другими словами $(Y_0 - \widehat{Y_0}) + (Y_1 - \widehat{Y_1}) + ... + (Y_n - \widehat{Y_n})$

$Y$ - фактические значения целевой переменной

$\widehat{Y}$ - предсказанные значения целевой переменной `модель.predict(предиктор/ы)`

$n$ - количество наблюдений

```python
Y_pr = model.predict(X)
MSE = sum((Y-Y_pr)**2)/len(Y)
MSE
```

Модель ошибается, нужно сравнить MSE c другими модели, чтобы выбрать лучшую

**Среднеабсолютная ошибка**:

$$ MAE = \frac{\Sigma |Y-\widehat{Y}| }{n} $$

$\Sigma$ - знак суммы, другими словами $(Y_0 - \widehat{Y_0}) + (Y_1 - \widehat{Y_1}) + ... + (Y_n - \widehat{Y_n})$

$Y$ - фактические значения целевой переменной

$\widehat{Y}$ - предсказанные значения целевой переменной `модель.predict(предиктор/ы)`

$n$ - количество наблюдений

```python
Y_pr = model.predict(X)
MAE = sum(abs(Y-Y_pr))/len(Y)
MAE
```

Модель ошибается, нужно сравнить MAE c другими модели, чтобы выбрать лучшую
